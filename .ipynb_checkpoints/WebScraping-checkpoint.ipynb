```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
URL ='https://listado.mercadolibre.com.co/ford-ecosport-titanium#D[A:ford%20ecosport%20titanium]'
page = requests.get(URL)
```

```python
html = BeautifulSoup(page.content, 'html.parser')
```

```python
results = html.find_all('ol',class_='ui-search-layout ui-search-layout--grid')
```

```python
links_of_publication=[]
for job_elem in results:
    links_publication=job_elem.find_all('a',class_='ui-search-result__content ui-search-link')
    for link in links_publication:
        links_of_publication.append(link.attrs['href'])
```

```python
links_of_publication
```

```python
def get_features(link):
    page = requests.get(link)
    html = BeautifulSoup(page.content, 'html.parser')
    
    rows_features = html.find('table',class_='andes-table')
    list_features_table = rows_features.find_all('tr',class_='andes-table__row')
    list_features={row.find('th').text: row.find('td').text for row in list_features_table}
    
    div_data_basica=html.find('div',class_='ui-pdp-container__row ui-pdp-component-list pr-16 pl-16')
    price=div_data_basica.find('span', class_='price-tag-fraction').text
    
    list_features.update({'Precio': price})
    
    return list_features
```

```python
def data_frame_features(links):
    list_features = [get_features(link) for link in links_of_publication]
    return pd.DataFrame(list_features)
```

```python
data_frame_features(links_of_publication)
```
